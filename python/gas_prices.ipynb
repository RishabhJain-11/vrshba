```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title("Gas Prices Overtime in USD")
plt.xlabel("Year of survey")
plt.ylabel("Price of Gas - USD")
# Plotting with labels
plt.plot(gas.Year, gas.USA, label="United States")
plt.plot(gas.Year, gas.UK, label="United Kingdom")
plt.plot(gas.Year, gas.Australia, label="Australia")
plt.plot(gas.Year, gas.Canada, label="Canada")

plt.xticks(gas.Year[::3])
# Displaying the legend
plt.legend()

# Displaying the plot
plt.show()
```

```python
fifa = pd.read_csv("fifa_data.csv")
fifa.head()
```

```python
plt.hist(fifa.Overall)
```

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count()[0]
right = fifa.loc[fifa["Preferred Foot"] == "Right"].count()[0]
labels = ["Left", "Right"]
colors = ["#abcdef", "#aabbcc"]
plt.pie([left, right], labels = labels, colors = colors, autopct = '%.2f %%')
plt.title("Foot Preference of FIFA Players")
plt.show()
```

```python
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
plt.style.use('ggplot')
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight > 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[(fifa.Weight >= 200)].count()[0]
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (.4,.2,0,0,.4)
plt.title("Weight Distribution of FIFA Players")
plt.pie(weights, labels = labels, autopct = "%.2f %%", pctdistance = 0.8, explode = explode)
plt.show
```

```python
plt.style.use('default')
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
mancity = fifa.loc[fifa.Club == 'Manchester City']['Overall']
arsenal = fifa.loc[fifa.Club == 'Arsenal']['Overall']
labels = ["FC Barcelona", "Real Madrid", "Man City","Arsenal"]
boxes = plt.boxplot([barcelona, madrid, mancity, arsenal], labels = labels, patch_artist = True, medianprops ={'linewidth' :2})

for box in boxes['boxes']:
    box.set(facecolor = "#4286f4", linewidth = 2)
    box.set(facecolor = "#e0e0e0")
plt.show()
```
